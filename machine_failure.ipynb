```python
import pandas as pd
import numpy as np
import sklearn 
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
X = data.drop(columns=["fail"]) 
y = data["fail"]  


clf = DecisionTreeClassifier(criterion='entropy') 


clf.fit(X, y)


plt.figure(figsize=(12,8))
tree.plot_tree(clf, feature_names=["footfall", "tempMode", "AQ", "USS", "CS", "VOC", "RP", "IP", "Temperature"], 
               class_names=["No", "Yes"], filled=True)
plt.show()
```

```python

prediction = clf.predict([[100, 3, 5, 2, 4, 2, 45, 5, 1]])


print(f"Predicted Fail: {'Yes' if prediction[0] == 1 else 'No'}")
```
